<h2 align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc"><a href="https://numpy.org/doc/stable/reference/routines.array-creation.html" target="_blank" style="color:#0099cc">🔗</a>ساخت آرایه
    </font>
</h2>

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    یکی از راه‌های ساخت آرایه نامپای ورودی دادن لیست پایتون به تابع <code>array</code> نامپای است.
    در کدهای زیر نمونه‌هایی از این نوع ایجاد آرایه را می‌بینید.
</font>
</p>

```python
# آرایه یک بعدی شامل 3 عدد
np.array([1, 2, 3])
```

```python
# آرایه دو بعدی 2*3 شامل 6 عدد 
np.array([[1, 2], 
         [3, 4],
         [5, 6]])
```

```python
# آرایه سه بعدی 2*2*2 شامل 8 رشته
np.array([[['000', '001'], ['010', '011']], 
          [['100', '101'], ['110', '111']]])
# dtype نوع خانه‌های آرایه را نشان می‌دهد
# <U3 که نوع این آرایه می‌شود به معنای رشته یونی‌کد 3 حرفی است 
```

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
یک راه دیگر برای ساخت آراه نامپای، استفاده از توابع داخلی نامپای است که چند نمونه از آن را در کدهای زیر می‌بینید.
    </font>
</p>

```python
# آرایه یک بعدی شامل 3 عدد حقیقی برابر صفر
np.zeros(3)
```

```python
# آرایه دو بعدی 3*2 شامل 6 عدد حقیقی برابر یک
np.ones((2, 3))
```

```python
# آرایه دو بعدی 2*3 شامل 6 رشته متنی '#' یکسان
np.full((3, 2), '#') 
```

```python
# آرایه یک بعدی شامل اعداد صحیح پیش از 15 با شروع از 0 و با فاصله 4
np.arange(0, 15, 4) 
```

```python
# آرایه یک بعدی شامل 5 عدد حقیقی در بازه 0 تا 20 (شامل این دو) با فاصله برابر
np.linspace(0, 20, 5) 
```

```python
# آرایه دو بعدی 3*3 متناظر با ماتریس همانی (قطر 1 و باقی خانه‌ها 0)
np.eye(3) 
```

```python
# آرایه یک بعدی شامل 3 عدد نامشخص (بسته به مقدار داخل حافظه)
np.empty(3) 
```

```python
# آرایه یک بعدی شامل 3 عدد تصادفی از توزیع یکنواخت بین 0 و 1
np.random.random(3) 
```

```python
# آرایه دو بعدی 3*3 شامل اعداد صحیح تصادفی بین 0 تا 9 
np.random.randint(0, 10, (3, 3)) 
```

```python
# آرایه دو بعدی 3*2 شامل اعداد تصادفی از توزیع نرمال با میانگین 0 و واریانس 1 
np.random.normal(0, 1, (2, 3)) 
```

```python
# آرایه دو بعدی 3*2 شامل اعداد صحیح تصادفی بین 0 تا 10 
a = np.random.randint(0, 10, (2, 3))
# آرایه پر از 3 با ابعاد مشابه آرایه ورودی (3*2)
np.full_like(a, 3) 
```

<h2 align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
نوع داده
    </font>
</h2>

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    در کدهای زیر، چگونگی تعیین نوع داده آرایه با تعیین <code>dtype</code> هنگام ساخت را می‌بینید.    </font>
</p>

```python
# نوع مقدار بولی
np.array([False, True], dtype='bool_') 
# کاراکتر _ برای اشاره به نوع داخل نامپای است اما می‌توان آن را برداشت
```

```python
# نوع عدد صحیح (عدد سوم تبدیل به 3 می‌شود)
np.array([1, 2, 3.14], dtype='int_') 
```

```python
# نوع عدد حقیقی 128 بیتی
np.array([2.71828182845, 3.14159265359], dtype='float128') 
```

```python
# نوع عدد صحیح نامنفی 8 بیتی
np.array([255], dtype='uint8') 
```

```python
# نوع رشته
np.array(['Hello', 'World'], dtype='str_') 
```

<h2 align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
مشخصه‌ها
    </font>
</h2>

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
در کدهای زیر، چند مثال از مشخصه‌های آرایه را مشاهده می‌کنید.
</p>

```python
a = np.array([[1, 2],
             [3, 4],
             [5, 6]], dtype='float64')
print(a)
print('number of dimensions:', a.ndim)  # تعداد ابعاد آرایه
print('shape:', a.shape)                # اندازه ابعاد آرایه
print('size:', a.size)                  # تعداد خانه‌های آرایه
print('dtype:', a.dtype)                # نوع داده آرایه
```

```python
c = np.array(['h', 'e', 'l', 'l', 'o'])
print("c is:",c)
print("c each item size is:",c.itemsize, 'bytes')   # تعداد بایت برای هر خانه آرایه
print("c total item size is:",c.nbytes, 'bytes')    # تعداد بایت برای کل آرایه
```
